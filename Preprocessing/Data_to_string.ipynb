
# Conversion of each column to a string separated by "_"

```python
data = pd.read_csv('../splited_alerts2019.csv', sep=';', 
                   header=0, index_col = 0)
```

```python
data['DATE_CASE'] =pd.to_datetime(data.DATE_CASE, yearfirst=True, format = '%Y/%m/%d')
datag = data
datag['DATE_CASE'] = data.DATE_CASE.dt.month
datag['DATE_CASE'] = datag['DATE_CASE'].astype(str)
datag['HAZARDS_CAT'] = datag['HAZARDS_CAT'].astype(str)
```

```python
datag.dtypes
```

```python
list1=[]
for i in data['COUNT_CONCERN']:

    if i == " ":
        list1.append(i)

    else:
        list1.append(i[:-1])
```

```python
data['COUNT_CONCERN'] = list1
```

```python
datag.NOT_COUNTRY = datag.NOT_COUNTRY.apply(lambda x: x.replace(' ','_'))
datag.PROD_CAT = datag.PROD_CAT.apply(lambda x: x.replace(' ','_'))
datag.RISK_DECISION = datag.RISK_DECISION.apply(lambda x: x.replace(' ','_'))
datag.ACTION_TAKEN = datag.ACTION_TAKEN.apply(lambda x: x.replace(' ','_'))
datag.DISTRIBUTION_STAT = datag.DISTRIBUTION_STAT.apply(lambda x: x.replace(' ','_'))
datag.HAZARDS_CAT = datag.HAZARDS_CAT.apply(lambda x: x.replace(' ','_'))
datag.COUNT_ORIGEN = datag.COUNT_ORIGEN.apply(lambda x: x.replace(' ','_'))
datag.TYPE = datag.TYPE.apply(lambda x: x.replace(' ','_'))
datag.COUNT_DESTIN = datag.COUNT_DESTIN.apply(lambda x: x.replace(' ','_'))
datag.COUNT_CONCERN = datag.COUNT_CONCERN.apply(lambda x: x.replace(' ','_'))
```
